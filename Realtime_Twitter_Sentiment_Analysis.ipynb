```python
import re

def preprocess_text(text):
    # convert to lower case
    text = text.lower()
    # remove user handle (Added 'r')
    text = re.sub(r"@[\w]*", "", text)
    # remove http links (Added 'r')
    text = re.sub(r"http\S+", "", text)
    # remove digits and spl characters (No backslash here, but 'r' is safe to add anyway)
    text = re.sub(r"[^a-zA-Z#]", " ", text)
    # remove rt characters
    text = re.sub(r"rt", "", text)
    # remove additional spaces (Added 'r')
    text = re.sub(r"\s+", " ", text)

    return text

# Install transformers if you haven't (restart runtime after)
# !pip install transformers scipy

from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
import numpy as np

# 1. Load a model trained specifically on Tweets (roberta)
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

def get_sentiment_roberta(text):
    # Roberta handles handles/urls better, but basic cleaning is still good
    encoded_input = tokenizer(text, return_tensors='pt')
    output = model(**encoded_input)

    # The model gives scores for [Negative, Neutral, Positive]
    scores = output[0][0].detach().numpy()
    scores = softmax(scores) # Convert to probabilities

    # Get the label with the highest score
    ranking = np.argsort(scores)
    top_label_index = ranking[-1]

    if top_label_index == 0:
        return "NEGATIVE"
    elif top_label_index == 1:
        return "NEUTRAL"
    else:
        return "POSITIVE"

# Test it on your failed example
tweet = "200,000+ users. Global hype. A new chapter begins for the $JBC ecosystem. üöÄüíé"
print(f"Sentiment: {get_sentiment_roberta(tweet)}")
```

```python
import time
import tweepy

# Make sure your client is set to wait automatically (if you can re-initialize it)
# client = tweepy.Client(bearer_token=..., wait_on_rate_limit=True)

print("Starting stream...")

while True:
    try:
        # 1. Get the response object first
        response = api.search_recent_tweets(query='#crypto', max_results=10)
        tweets = response.data

        # 2. Check if tweets exist (tweets is None if nothing is found)
        if tweets is not None:
            for tweet in tweets:
                original_tweet = tweet.text
                clean_tweet = preprocess_text(original_tweet)
                sentiment = get_sentiment(clean_tweet)

                print('------------------------Tweet-------------------------------')
                print(original_tweet)
                print('------------------------------------------------------------')
                print('Sentiment:', sentiment)
                print('\n')
        else:
            print("No tweets found for this query right now.")

        # 3. CRITICAL: Sleep here to pause before the next API Request
        # Twitter Basic Tier has a limit (e.g., 60 requests per 15 mins).
        # Waiting 15-20 seconds is safer.
        print("Waiting 20 seconds before next request...")
        time.sleep(20)

    except tweepy.TooManyRequests:
        print("Rate limit hit! Sleeping for 15 minutes...")
        time.sleep(15 * 60)

    except Exception as e:
        print(f"An error occurred: {e}")
        time.sleep(10)
```
